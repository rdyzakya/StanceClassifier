```python
import os
import json
base_path = os.path.join("..","mono")
paths = [os.path.join(base_path,el) for el in os.listdir(base_path)]
```

```python
paths
```

```python
combined_data = []

for p in paths:
    data_path = os.path.join(p,"interim","data.txt")
    label_mapping_path = os.path.join(p,"interim","label_mapping.json")

    with open(data_path,"r") as f:
        data = f.read().split("\n\n")
    data = [[line.split('\t') for line in sen.splitlines()] for sen in data]
    
    with open(label_mapping_path,"r") as f:
        label_mapping = json.load(f)
    
    for i_sen in range(len(data)):
        for i_line in range(len(data[i_sen])):
            # uncased token
            data[i_sen][i_line][0] = data[i_sen][i_line][0].lower()
            data[i_sen][i_line][1] = label_mapping[data[i_sen][i_line][1]]
    
    combined_data.extend(data)
```

```python
combined_data
```

```python
# write data to data.txt
with open("data.txt","w") as f:
    for sen in combined_data:
        for line in sen:
            f.write("\t".join(line))
            f.write("\n")
        f.write("\n")
```

```python
labels = set()
for sen in combined_data:
    for line in sen:
        labels.add(line[1])
```

```python
labels
```

```python
# sort
labels = sorted(list(labels))
# write to labels.txt
with open("labels.txt","w") as f:
    f.write("\n".join(labels))
```
